```python
from pyvis.network import Network
import networkx as nx

# Create a directed graph
graph = nx.DiGraph()

# Add nodes with labels
graph.add_node("Albert Einstein", label="Albert Einstein")
graph.add_node("Isaac Newton", label="Isaac Newton")
graph.add_node("Marie Curie", label="Marie Curie")
graph.add_node("Scientific Achievements", label="Scientific Achievements")

# Add edges with labels
graph.add_edge("Albert Einstein", "Ulm, Germany", label="Was born in")
graph.add_edge("Albert Einstein", "Theory of Relativity", label="Developed")
graph.add_edge("Isaac Newton", "Woolsthorpe, England", label="Was born in")
graph.add_edge("Isaac Newton", "Laws of Motion", label="Developed")
graph.add_edge("Marie Curie", "Warsaw, Poland", label="Was born in")
graph.add_edge("Marie Curie", "Radioactivity", label="Discovered")
graph.add_edge("Albert Einstein", "Scientific Achievements", label="Contributed to")
graph.add_edge("Isaac Newton", "Scientific Achievements", label="Contributed to")
graph.add_edge("Marie Curie", "Scientific Achievements", label="Contributed to")

# Initialize Pyvis network
nt = Network("1000px", "800px", notebook=True, heading='Science', cdn_resources='in_line')

# Import the networkx graph
nt.from_nx(graph)

# Set physics and configure options
nt.set_options("""
{
  "nodes": {
    "font": {
      "size": 16
    }
  },
  "edges": {
    "font": {
      "size": 12
    },
    "arrows": {
      "to": {
        "enabled": true,
        "scaleFactor": 1.2
      }
    }
  },
  "physics": {
    "enabled": true
  },
  "interaction": {
    "hover": true
  }
}
""")


# Generate and display the HTML content directly
nt.show("graph.html")
with open("graph.html", "r") as file:
    html_content = file.read()
display(HTML(html_content))
```

## **KG data defined by pandas DataFrames**



```python
import pandas as pd
from pyvis.network import Network

# Define node data
node_data = {
    'id': ['Vendor', 'Carrier', 'Shipment1', 'Shipment2', 'Inventory1', 'Inventory2', 'Product', 'ProductGroup', 'Delivery1', 'Customer1', 'Customer2', 'Carrier2'],
    'label': ['Vendor', 'Carrier', 'Shipment', 'Shipment', 'Inventory', 'Inventory', 'Product', 'Product Group', 'Delivery', 'Customer', 'Customer', 'Carrier'],
    'title': [
        "company_name: Acme Manufacturing\ncity: Macon\nstate: GA",
        "company_name: UPS",
        "shipment_id: 3233\nship_date: 2023-5-3\nreceive_date: 2023-5-10",
        "shipment_id: 234\nship_date: 2023-6-1\nreceive_date: 2023-6-3",
        "warehouse_name: East Lansing\nproduct_id: 5432",
        "warehouse_name: Grand Rapids\nproduct_id: 5432",
        "product_name: Velvet sofa\nproduct_id: 5432",
        "group_name: Living room furniture\ngroup_id: 15",
        "delivery_id: 123\ndelivery_date: 2023-8-1",
        "name: John Doe",
        "name: Jane Doe",
        "company_name: FedEx"
    ],
    'color': ['#8000ff', '#ffeb3b', '#ffffff', '#ffffff', '#00ff00', '#00ff00', '#00d1b2', '#000000', '#ff5722', '#8000ff', '#8000ff', '#ffeb3b']
}

# Define edge data
edge_data = {
    'source': ['Vendor', 'Shipment1', 'Carrier', 'Shipment2', 'Shipment1', 'Inventory1', 'Inventory2', 'Product', 'Inventory2', 'Delivery1', 'Inventory1', 'Carrier', 'Shipment2', 'Carrier2', 'Delivery1'],
    'target': ['Shipment1', 'Carrier', 'Shipment2', 'Inventory1', 'Inventory1', 'Product', 'Product', 'ProductGroup', 'Delivery1', 'Customer1', 'Inventory2', 'Shipment1', 'Carrier2', 'Inventory2', 'Customer2'],
    'label': ['SUPPLIES', 'DELIVERS', 'DELIVERS', 'RECEIVES', 'RECEIVES', 'CONTAINS', 'CONTAINS', 'CATEGORIZES', 'FULFILLS', 'SHIPS TO', 'PULLS', 'DELIVERS', 'DELIVERS', 'DELIVERS', 'SHIPS TO']
}

# Convert dictionaries to DataFrames
nodes_df = pd.DataFrame(node_data)
edges_df = pd.DataFrame(edge_data)

# Initialize Pyvis network
nt = Network("1000px", "800px", notebook=True, heading='Graph Model: Supply Chain', cdn_resources='in_line')

# Add nodes from DataFrame
for index, row in nodes_df.iterrows():
    nt.add_node(row['id'], label=row['label'], title=row['title'], color=row['color'])

# Add edges from DataFrame
for index, row in edges_df.iterrows():
    nt.add_edge(row['source'], row['target'], label=row['label'])

# Set options for visualization
nt.set_options("""
{
  "nodes": {
    "font": {
      "size": 16
    }
  },
  "edges": {
    "font": {
      "size": 12
    },
    "arrows": {
      "to": {
        "enabled": true,
        "scaleFactor": 1.2
      }
    }
  },
  "physics": {
    "enabled": true
  },
  "interaction": {
    "hover": true
  }
}
""")



# Generate and display the HTML content directly
nt.show("graph.html")
with open("graph.html", "r") as file:
    html_content = file.read()
display(HTML(html_content))

```
